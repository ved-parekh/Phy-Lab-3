```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Load the CSV file
df1=pd.read_excel('coupled pendulum data.xlsx', sheet_name='green inphase', skiprows=[1])
data1=df1.to_numpy()

# mass b in sheetname 'green in phase'
t1=df1.iloc[:,0]
x1=df1.iloc[:,1]

time1=t1.to_numpy() # x axis
amp1=x1.to_numpy() # y axis

t2=df1.iloc[:,4]
x2=df1.iloc[:,5]

time2=t2.to_numpy()
amp2=x2.to_numpy()

# Compute sampling interval (T) and number of samples (N)
T = np.mean(np.diff(time1))  # Average time difference
N = len(time1)  # Number of points

# Perform FFT
yf = fft(amp1)
yf[0]=0
xf = fftfreq(N, T)[:N//2]  # Get positive frequencies

max_amp=np.argmax(yf)
print(xf[max_amp])

diff=np.max(amp1) - np.min(amp1)
norm_amp1= (amp1 - np.min(amp1))/diff

diff2=np.max(amp2) - np.min(amp2)
norm_amp2= (amp2 - np.min(amp2))/diff2

plt.plot(time2[:633],norm_amp2[:633])
plt.plot(time1[:633], norm_amp1[:633])
plt.xlabel('time (s)')
plt.ylabel('position (cm)')

# Plot FFT
plt.figure(figsize=(10, 5))
plt.plot(xf, 2.0/N * np.abs(yf[:N//2]))  # Normalize amplitude
plt.grid()
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.title("FFT of green spring in phase (mass A)")
plt.xlim(right=4)
plt.xlim(left=0)
plt.show()


```
