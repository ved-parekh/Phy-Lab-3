```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
#we are trying to obtain y from interpolating x, x=refractive index, y=corresponding wavelength
#x = refractive index, y=wavelength

# Given x-axis and y-axis data
x_values = np.array([
    1.583536571, 1.583384423, 1.583231675, 1.583078323, 1.582924364,
    1.582769792, 1.582614604, 1.582458797, 1.582302365, 1.582145305,
    1.581987613,1.585317139, 1.600035957, 1.610293883, 1.646092474, 1.668734299, 1.696015997
])

y_values = np.array([
    2.4, 2.41, 2.42, 2.43, 2.44, 2.45, 2.46, 2.47, 2.48, 2.49, 2.5, 2.28, 1.17 ,0.75, 0.42,0.36, 0.32
])

# Create interpolation function (cubic for smoothness)
interp_func = interpolate.interp1d(x_values, y_values, kind='cubic')

# Generate new x values for interpolation
x_new = np.linspace(min(x_values), max(x_values), 218)

# Compute interpolated y values
y_new = interp_func(x_new)

# Plot original data and interpolated curve
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, 'o', label="Original Data", markersize=6)
plt.plot(x_new, y_new, '-', label="Interpolated Curve")
plt.xlabel("refractive index")
plt.ylabel("wavlength")
plt.title("Interpolation Using the Chart")
plt.legend()
plt.grid()
plt.show()
```
