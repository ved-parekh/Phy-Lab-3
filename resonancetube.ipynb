### 1st harmonic


```python
import numpy as np
import matplotlib.pyplot as plt

recip_freq= [0.001953125, 0.002083333333,0.002344116268,0.002604166667,0.00292997363,0.003125,0.003472222222,0.00390625]

lambda1= [0.62,0.66,0.72,0.81,0.93,0.98,1.13,1.27]

lambda2= [0.62,0.69,0.78,0.87,0.99,1.06,1.16,1.33]


fit=np.polyfit(lambda1, recip_freq,1)
bestfitline=np.polyval(fit,lambda1)

slope=fit[0]
intercept=fit[1]

print(f"Best fit line equation: y = {slope:}x + {intercept:}")


plt.plot(lambda1, bestfitline,color='red',ls='--')
plt.scatter(lambda1, recip_freq)
```
