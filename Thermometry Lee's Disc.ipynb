```python
import numpy as np
import matplotlib.pyplot as plt

time=[0,42,78,132,184,229,273,330,384,447,496,561,617,684,744,817,883,952,1005,1090,1171]
temp= [52,51.5,51,50.5,50,49.5,49,48.5,48,47.5,47,46.5,46,45.5,45,44.5,44,43.5,43,42.5,42]

plt.scatter(time, temp)

m,c=np.polyfit(time,temp,1)

bestfitline=[-0.0087*x+ c for x in time]

# bestfitline2= [m*x + c for x in time] 

plt.plot(time, bestfitline , linestyle='dashed',  color='red', label='Best fit line ')
# plt.plot(time, bestfitline2 , linestyle='dashed',  color='red', label='Best fit line ')


#plt.axhline(y=47, linestyle= 'dashed', color= 'black')
plt.xlabel('Time (s)')
plt.ylabel('Temperature ($\degree C$)')
plt.legend()
print(f"slope: {m}")
print(f"intercept: {c}")
```
